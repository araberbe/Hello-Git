**Configurar SSH**

Protocolo de red seguro utilizado para acceder y gestionar sistemas de manera remota. Permite que un usuario se conecte a otro servidor o dispositivo a trav√©s de una red (como Internet) de forma segura, cifrando la comunicaci√≥n entre ambos

1) Generar clave SSH: **`ssh-keygen -t ed25519 -C "tu_correo@example.com"`**
2) Clave p√∫blica:**`cat ~/.ssh/id_ed25519.pub`**
3) Agregar clave p√∫blica a GitHub/GitLab > Settings > SSH and GPG Keys
4) Verificar conexi√≥n : **`ssh -T git@github.com`**
5) Iniciar el agente SSH: **`eval (ssh-agent -s)`**         
6) Agregar la clave al agente SSH: **`ssh-add ~/.ssh/id_ed25519`**
